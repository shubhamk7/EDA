```python
def Zscore(x,mean_x,std_dev_x):
    return (x-mean_x)/std_dev_x
```

```python
mean_x = df1.Price.mean()
std_dev_x = df1.Price.std()
for x in df1.Price:
    print('/n Zscore is:',Zscore(x,mean_x,std_dev_x))
```

```python
for x in set(df1['Item']):
    if x=='Jeans':
        df1['Item']=df1['Item'].replace(x,'1')
    elif x=='Trousar':
        df1['Item']=df1['Item'].replace(x,'2')
    elif x=='Formal ':
        df1['Item']=df1['Item'].replace(x,'3')
```

```python
for x in set(df1['Franchise']):
    if x=='Hyderabad':
        df1['Franchise']=df1['Franchise'].replace(x,'1')
    elif x=='Pune':
        df1['Franchise']=df1['Franchise'].replace(x,'2')
    elif x=='Chennai':
        df1['Franchise']=df1['Franchise'].replace(x,'3')
    elif x=='US':
        df1['Franchise']=df1['Franchise'].replace(x,'4')
```

```python
for x in set(df1['Payment_Mode']):
    if x=='Debit Card':
        df1['Payment_Mode']=df1['Payment_Mode'].replace(x,'1')
    elif x=='Credit Card':
        df1['Payment_Mode']=df1['Payment_Mode'].replace(x,'2')
    elif x=='Cash Payment':
        df1['Payment_Mode']=df1['Payment_Mode'].replace(x,'3')
```

```python
for x in set(df1['Gender']):
    if x=='Male':
        df1['Gender']=df1['Gender'].replace(x,'0')
    elif x=='Female':
        df1['Gender']=df1['Gender'].replace(x,'1')
```

```python
 def getEValue(column_name):
        df_test = pd.DataFrame(pd.crosstab(newDf[column_name],newDf.target,margins=True))
        df_r_All = df_test.ix['All']
        df_c_All = df_test['All']
        sample = df_r_All.ix['All']
        for x in range(len(df_r_All.index.values)-1):
            return(df_r_All.ix[x]*df_c_All[x])/sample
```

```python
for x in newDf.columns:
    if x!='target':
        print(x,':',getEValue(x))
```
